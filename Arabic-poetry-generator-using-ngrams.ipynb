```python
# Omar Alkhuwaytim 440013412 
import nltk
from nltk import ngrams
import random
from nltk.corpus import PlaintextCorpusReader


# Load a corpus of text data
corpus_root = 'Corpus\Corpus' 
```

```python
# tokenization
from nltk.tokenize import word_tokenize

with open('cb1.txt', 'r', encoding='utf-8') as f:
    contents = f.read()
    
contents_tokens = nltk.word_tokenize(contents)

print(contents_tokens)
    
```

```python
# Create the ngrams
from nltk.util import ngrams
# 5 = number of grams
contents_ngrams = list(nltk.ngrams(contents_tokens, 5))

n_grams = {}
for sentence in contents_ngrams:
    words = [word for word in sentence if word[0].isalpha()]
    for ix in range(len(words) - 1):
        try:
            n_grams[words[ix]].append(words[ix + 1])
        except KeyError as _:
            n_grams[words[ix]] = []
            n_grams[words[ix]].append(words[ix + 1])
```

```python
# generate the sentences 
# nb = sentence lengthe
import random
def generate_sentence(nb , fw):
    words = []
    next_word = random.choice(list(n_grams.keys()))
    words.append(fw)
    while len(words) < nb:
        next_word = random.choice(n_grams[next_word])
        words.append(next_word)
        
    return " ".join(words)
```
